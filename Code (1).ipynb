After importing and inspecting the price dataset, I observed a large number of missing values across stocks, as well as some zero and infinite price entries, which are not economically meaningful. Rather than attempting to interpolate or forward-fill prices—which would artificially smooth the series—or dropping entire stocks, which would discard useful information, I followed a conservative approach. I converted all zero and infinite values to NaN and computed returns first, because return outliers are more interpretable and far less ambiguous than price outliers. Once returns were calculated, I applied a robust outlier-filtering step using the Median Absolute Deviation (MAD) with a threshold of 5 to remove only clearly erroneous observations without distorting genuine market moves. This approach avoids fabricating data, preserves the natural structure of the series, and is more appropriate than alternatives such as winsorization, linear interpolation, or σ-based filtering, all of which can materially alter return distributions or bias subsequent analyses.

Outlier removal (this step is done post returns calculation )

```python
def remove_outliers_mad(r, threshold=5):
    median = r.median()
    mad = (abs(r - median)).median()
    
    # If MAD is zero, cannot compute — return unchanged
    if mad == 0 or np.isnan(mad):
        return r

    modified_z = 0.6745 * (r - median) / mad
    
    # Replace outliers with NaN
    return r.mask(abs(modified_z) > threshold)
```

```python
returns = returns.apply(remove_outliers_mad, axis=1)
```

```python
returns.stack()
```

```python
alpha=0.05
```

```python
results=[]
```

```python
from scipy.stats import jarque_bera, skew, kurtosis
```

```python
for stock in returns.index:
    # print(returns.loc[stock])
    r = returns.loc[stock].dropna()
    n = len(r)
    jb_stat, p_value = jarque_bera(r)
    sk = float(skew(r, bias=False))
    kt = float(kurtosis(r, fisher=False, bias=False))
    normal_flag = bool(p_value >= alpha)
    
    results.append({
        "Stock": stock,
        "n_returns": n,
        "JB_stat": float(jb_stat),
        "p_value": float(p_value),
        "skewness": sk,
        "kurtosis": kt,
        "normal_at_0.05": normal_flag,
        "note": ""
    })
    
```

```python
results_df= pd.DataFrame(results)
```

Jarque–Bera tests show that most stock return series strongly reject normality, confirming that returns are not normally distributed. This is consistent with financial data, where returns typically exhibit skewness, excess kurtosis, and fat tails. The lack of normality is important because many classical models—such as CAPM, mean–variance portfolio optimisation, and the Black–Scholes option pricing framework—use normality mainly as a simplifying assumption. When returns are heavily non-normal, these models can misestimate risk, produce unstable parameter estimates, and underestimate the probability of extreme outcomes. That said, they are still widely used in practice, but their outputs must be interpreted carefully and often supplemented with more robust techniques, such as shrinkage estimators, implied volatility, or heavy-tailed distributions.

```python
for stock in returns.index[:50]:
    plt.figure(figsize=(4,4))
    stats.probplot(returns.loc[stock].dropna(), dist="norm", plot=plt)
    plt.title(f"Q-Q Plot: {stock}")
    plt.show()
```

The Q–Q plots further confirm that the stock returns are not normally distributed. Instead of lying on the 45° reference line, the empirical quantiles exhibit a pronounced S-shape, with the tails bending away from the line. This indicates strong excess kurtosis and heavy tails, meaning extreme positive and negative returns occur more frequently than a normal distribution predicts. Combined with the Jarque–Bera tests, the Q–Q plots provide clear visual evidence of significant departures from normality in the return series.

```python


import pandas as pd
import numpy as np
import statsmodels.api as sm

# returns_clean: rows = stocks, columns = dates
# market_index: Series indexed by same dates (columns of returns_clean)

betas = []
alphas = []
stock_ids = []
r2_list = []
std = []
residual_std = []
```

```python
for stock in returns.index:

    # stock's cleaned return series
    r_i = returns.loc[stock]

    # align returns with market index
    df_xy = pd.concat([r_i, market_index], axis=1)
    df_xy.columns = ["ri", "rm"]

    # remove NaNs
    df_xy = df_xy.dropna()

    if len(df_xy) < 20:   # require enough data points
        betas.append(np.nan)
        alphas.append(np.nan)
        r2_list.append(np.nan)
        stock_ids.append(stock)
        continue

    # regression: ri = alpha + beta * rm
    X = sm.add_constant(df_xy["rm"])
    y = df_xy["ri"]

    model = sm.OLS(y, X).fit()

    resid = model.resid  # residual series
    std_r = df_xy["ri"].std(ddof=1)           # sample standard deviation
    std_resid = resid.std(ddof=1)

    alphas.append(model.params["const"])
    betas.append(model.params["rm"])
    r2_list.append(model.rsquared)
    stock_ids.append(stock)
    std.append(std_r)
    residual_std.append(std_resid)
```

Question 9- Code for calculating rolling Betas

```python

WINDOW = 250   # 250-day rolling window

# Make sure data aligns
market = market_index.loc[returns.columns]

# Prepare a DataFrame to store rolling betas
rolling_betas = pd.DataFrame(index=returns.index, columns=returns.columns)

for stock in returns.index:

    r_i = returns.loc[stock]
    # print(stock)

    # Loop over all possible end dates
    for t in range(WINDOW, len(returns.columns)):

        # Select 250-day window
        window_r_i = r_i.iloc[t-WINDOW:t]
        window_rm  = market.iloc[t-WINDOW:t]

        # Drop NaNs
        df_xy = pd.concat([window_r_i, window_rm], axis=1).dropna()
        df_xy.columns = ["ri","rm"]

        # Require sufficient data in the window
        if len(df_xy) < 50:
            continue

        X = sm.add_constant(df_xy["rm"])
        y = df_xy["ri"]

        model = sm.OLS(y, X).fit()
        rolling_betas.iloc[rolling_betas.index.get_loc(stock), t] = model.params["rm"]
```

Question 10 

Rolling stock-level betas provide a useful backward-looking measure of systematic risk, but they are not very reliable predictors of future beta. Individual stock betas are inherently noisy because idiosyncratic volatility dominates market-related variation, leading to high estimation error even with long windows such as 250 days. In practice, a company’s risk exposure changes over time due to shifts in leverage, business models, industry conditions, and market regimes, making historical betas slow to adjust. As a result, raw rolling betas tend to be unstable and exhibit weak out-of-sample predictive power

Question 11

A more robust estimate of future betas can be obtained by reducing the noise and instability inherent in raw rolling-window regressions. One improvement is to apply shrinkage (e.g., Blume or Vasicek), which pulls noisy estimates toward a long-run or industry-level beta. Another approach is to estimate portfolio or sector betas, which diversify idiosyncratic noise. Using time-varying models can also help: for example, computing beta using EWMA (Exponentially Weighted Moving Averages) gives more weight to recent data and adapts more smoothly to changing market regimes, improving out-of-sample stability.
